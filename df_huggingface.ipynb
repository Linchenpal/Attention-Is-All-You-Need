```python


# Function to process the dataset in chunks
def get_huggingface_texts(number):
    path = "model_training_dataset.csv"
    chunksize = number  # Set chunk size, e.g., 10000 for smaller chunks

    # Initialize empty lists to store chunked data
    human_text_data = []
    ai_text_data = []

    for chunk in pd.read_csv(path, chunksize=chunksize):
        # Process each chunk separately

        # Human text chunk
        human_chunk = chunk[['id', 'human_text', 'instructions']].copy()
        human_chunk.rename(columns={'human_text': 'text'}, inplace=True)
        human_chunk['generated'] = 0  # Label as human-generated
        human_text_data.append(human_chunk)

        # AI text chunk
        ai_chunk = chunk[['id', 'ai_text', 'instructions']].copy()
        ai_chunk.rename(columns={'ai_text': 'text'}, inplace=True)
        ai_chunk['generated'] = 1  # Label as AI-generated
        ai_text_data.append(ai_chunk)

    # Concatenate all chunks after processing
    huggingface_combined_text_df = pd.concat(human_text_data + ai_text_data, ignore_index=True)

    return huggingface_combined_text_df

# Load and process the data in chunks, with each chunk size of 10,000 rows
huggingface_combined_text_df = get_huggingface_texts(10000)
huggingface_combined_text_df.head()
```

# Text-Based Features

```python
def text_based_features(df, text_column):
    # Word count
    df[f'{text_column}_word_count'] = df[text_column].apply(lambda x: len(str(x).split()))

    # Character count
    df[f'{text_column}_char_count'] = df[text_column].apply(lambda x: len(str(x)))

    # Average word length
    df[f'{text_column}_avg_word_length'] = df[f'{text_column}_char_count'] / df[f'{text_column}_word_count']

    # Sentence count
    df[f'{text_column}_sentence_count'] = df[text_column].apply(lambda x: len(sent_tokenize(str(x))))

    # Average sentence length
    df[f'{text_column}_avg_sentence_length'] = df[f'{text_column}_word_count'] / df[f'{text_column}_sentence_count']

    return df
```

# Lexical Diversity and Readability Scores

```python
def lexical_readability_features(df, text_column):
    # Lexical Diversity
    df[f'{text_column}_lexical_diversity'] = df[text_column].apply(lambda x: len(set(str(x).split())) / len(str(x).split()))

    # Flesch Reading Ease
    df[f'{text_column}_flesch_reading_ease'] = df[text_column].apply(lambda x: Readability(str(x)).flesch().score)

    # Gunning Fog Index
    df[f'{text_column}_gunning_fog'] = df[text_column].apply(lambda x: Readability(str(x)).gunning_fog().score)

    return df
```
